```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#going through xlsx file. This has multiple datasets. 

def reindex_sheet(sheet_name, skiprownumber, column_name): #functions will streamline the process as we have multiple sheets to reindex.
    skip_rows = skiprownumber
    df1 = pd.read_excel("../datasets/master_productivityds.xlsx", sheet_name=sheet_name, skiprows=skip_rows)
    
    df1.rename(
        columns={'Activities of head offices; management consultancy activities': 'Management Consulting'}, 
        inplace=True
    )

    isolated_df = df1[['Quarter', 'Management Consulting']].copy()
    #changing from index=2023 to index=2025
    val_2023 = isolated_df.loc[isolated_df['Quarter'] == '2023 Q1', 'Management Consulting'].values[0]
    val_2025 = isolated_df.loc[isolated_df['Quarter'] == '2025 Q1', 'Management Consulting'].values[0]
    scaling_factor = val_2025 / val_2023 * 100
    
    isolated_df.loc[:, column_name] = (
        isolated_df['Management Consulting'] * scaling_factor / 100
    )
    
    reindexed_df = isolated_df.drop('Management Consulting', axis=1)
    return reindexed_df

#our dataframes
GVA2025_table23 = reindex_sheet("Table_23", [0,1,2,4,5,6], 'Gross Added Value')
HoursWorked2025_table25 = reindex_sheet("Table_25", [0,1,2,3,5,6,7], 'Hours Worked')
HourlyOutput2025_table27 = reindex_sheet("Table_27", [0,1,2,3,5,6,7], 'Hourly Output')

```

```python
#basic plot - the below took some AI assistance just for time efficiency. It is not the EDA or final graph, just to test that everything has been rendered properly.
plt.figure()
plt.plot(
    GVA2025_table23['Quarter'].iloc[:20],
    GVA2025_table23['Gross Added Value'].iloc[:20]
)
plt.xticks(rotation=45)
plt.xlabel('Quarter')
plt.ylabel('Gross Added Value (Index 2025)')
plt.title('Management Consulting GVA (Index 2025)  First 20 Quarters')
plt.tight_layout()
plt.show()
```
