```python
#All imports below.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cleaning our first dataset, our management consulting. Just want data on a quarterly basis...
pd.set_option('display.max_rows', None)
rev_df = pd.read_csv("datasets/total_revenue.csv", skiprows=0)
rev_df = rev_df.drop(rev_df.index[0:34]).reset_index(drop=True)
rev_df = rev_df.iloc[:111]
rev_df.rename(columns={'Title': 'Quarter', 'GBServTO: 70.2 - Management consulting services TOTAL (£m)': 'Management Consulting Services Total (£m)'}, inplace=True)
rev_df['Management Consulting Services Total (£m)'] = pd.to_numeric(rev_df['Management Consulting Services Total (£m)'], errors='coerce')
rev_df['Quarter'] = rev_df['Quarter'].astype(str)
print(rev_df)
```

```python
subset = rev_df.iloc[20:50]

#Basic line plot - the below took some AI assistance just for time efficiency. It is not the EDA or final graph, just to test the type allocation is working.
plt.figure(figsize=(10,5))
plt.plot(subset['Quarter'], subset['Management Consulting Services Total (£m)'], marker='o')
plt.title('Management Consulting Services Total (£m) for First 20 Quarters')
plt.xlabel('Quarter')
plt.ylabel('Total (£m)')
plt.xticks(rotation=45) 
plt.grid(True)
plt.tight_layout()
plt.show()
```
